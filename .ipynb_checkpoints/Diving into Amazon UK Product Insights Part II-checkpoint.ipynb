```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# 1. Load the data
path = r'D:\Ironhack\Week5\Day1\amz_uk_price_prediction_dataset.csv'
df = pd.read_csv(path)

# --- CROSSTAB ANALYSIS ---

# Create the frequency table
ct = pd.crosstab(df['category'], df['isBestSeller'])

# Calculate the proportion of best-sellers per category
# We add total for the calculation, then sort by proportion
ct['total'] = ct[True] + ct[False]
ct['proportion'] = ct[True] / ct['total']
ct_sorted = ct.sort_values(by='proportion', ascending=False)

print("--- Top 5 Categories with highest Best-Seller prevalence ---")
print(ct_sorted[['proportion']].head())
print("-" * 50)

# --- STATISTICAL TESTS ---

# Chi-square test (on the raw counts, not proportions)
chi2_data = pd.crosstab(df['category'], df['isBestSeller'])
chi2, p, dof, expected = chi2_contingency(chi2_data)

# Cramér's V Calculation
n = chi2_data.sum().sum()
# min_dim is (number of categories - 1) or (2 statuses - 1), whichever is smaller
min_dim = min(chi2_data.shape) - 1
cramers_v = np.sqrt(chi2 / (n * min_dim))

print(f"Chi-square Statistic: {chi2:.2f}")
print(f"P-value: {p:.4e}")  # Scientific notation for very small values
print(f"Cramér's V (Strength of Association): {cramers_v:.4f}")
print("-" * 50)

# --- VISUALIZATION ---

# Prepare data for plotting: Take top 10 categories by proportion
plot_data = ct_sorted.head(10).copy()

# Rename columns to strings to prevent Pandas keyword errors
plot_data.columns = [str(col) for col in plot_data.columns]

# Create Stacked Bar Chart
plot_data[['False', 'True']].plot(kind='bar', 
                                  stacked=True, 
                                  figsize=(12, 6), 
                                  color=['#3498db', '#e74c3c'])

plt.title('Distribution of Best-Sellers across Top 10 Categories', fontsize=14)
plt.xlabel('Product Category', fontsize=12)
plt.ylabel('Count of Products', fontsize=12)
plt.legend(['Not Best-Seller', 'Best-Seller'], title='Status', loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Business Logic Summary for your Report:

Prevalence: The printed ct_sorted table identifies which categories have the highest conversion rate to "Best-Seller" status. If the proportion is high, it suggests a category where the "Best-Seller" badge is more attainable or highly concentrated.Significance: With a $P$-value usually much lower than $0.05$, you can confirm that Best-Seller status is not distributed randomly—the category a product belongs to significantly influences its chances of being a best-seller.Strength: The Cramér's V tells you the magnitude. A value near $0.1$ is a "small" effect, while $0.3$ is "medium." In retail, even a small effect is useful for deciding where to launch new products.

```python
# --- PART 2: PRICES AND RATINGS ---

# 1. Preliminary Step: Remove Outliers using IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create the cleaned dataframe
df_clean = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)].copy()

print(f"Outlier Removal Summary:")
print(f"Lower Bound: £{lower_bound:.2f}, Upper Bound: £{upper_bound:.2f}")
print(f"Remaining products: {len(df_clean)} (Removed {len(df) - len(df_clean)} outliers)")
print("-" * 50)

# 2. Violin Plots: Price Distribution
# Filtering for the top 20 categories by count for better visualization
top_20_cats = df_clean['category'].value_counts().nlargest(20).index
df_top20 = df_clean[df_clean['category'].isin(top_20_cats)]

plt.figure(figsize=(16, 8))
sns.violinplot(x='category', y='price', data=df_top20)
plt.title('Price Distribution across Top 20 Categories (Outliers Removed)', fontsize=14)
plt.xticks(rotation=90)
plt.show()

# Highest Median Price (on the full cleaned dataset)
medians = df_clean.groupby('category')['price'].median()
print(f"Category with the highest median price: {medians.idxmax()} (£{medians.max():.2f})")

# 3. Bar Charts: Average Price (Top 10 Categories)
top_10_cats = df_clean['category'].value_counts().nlargest(10).index
df_top10 = df_clean[df_clean['category'].isin(top_10_cats)]

plt.figure(figsize=(12, 6))
df_top10.groupby('category')['price'].mean().sort_values(ascending=False).plot(kind='bar', color='teal')
plt.title('Average Price of Top 10 Most Frequent Categories', fontsize=14)
plt.ylabel('Average Price (£)')
plt.xticks(rotation=45)
plt.show()

# Highest Average Price overall
means = df_clean.groupby('category')['price'].mean()
print(f"Category with the highest average price: {means.idxmax()} (£{means.max():.2f})")

# 4. Box Plots: Ratings (Top 10 Categories)
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='stars', data=df_top10)
plt.title('Distribution of Product Ratings (Top 10 Categories)', fontsize=14)
plt.xticks(rotation=45)
plt.show()

# Highest Median Rating overall
median_ratings = df_clean.groupby('category')['stars'].median()
print(f"Category with the highest median rating: {median_ratings.idxmax()} ({median_ratings.max()} stars)")
```

Part 2: Pricing and Rating Analysis
1. Data Cleaning: IQR Outlier Removal
Before analyzing prices, we must remove extreme values. In retail datasets, high-end luxury items or pricing errors can pull the mean away from the typical customer experience.

```python
# Calculate Q1, Q3 and IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataframe
df_clean = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)].copy()

print(f"Original items: {len(df)}")
print(f"Items after outlier removal: {len(df_clean)}")
```

2. Violin Plots: Price Distribution
Violin plots allow us to see the "density" of prices. A wider section of the "violin" indicates a price point where most products in that category are clustered.

```python
# Top 10 categories by count
top_10_counts = df_clean['category'].value_counts().nlargest(10).index
df_top10 = df_clean[df_clean['category'].isin(top_10_counts)]

plt.figure(figsize=(12, 6))
df_top10.groupby('category')['price'].mean().sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Average Price of Top 10 Most Frequent Categories')
plt.ylabel('Average Price (£)')
plt.show()

# Highest Average Price (No filter)
highest_avg_cat = df_clean.groupby('category')['price'].mean().idxmax()
highest_avg_val = df_clean.groupby('category')['price'].mean().max()
print(f"The category with the highest average price is: {highest_avg_cat} (£{highest_avg_val:.2f})")
```

4. Box Plots: Customer Ratings
Finally, we analyze the stars column. Box plots are perfect here because they show the range of customer satisfaction and identify categories where ratings are consistently high versus those that are volatile.

```python
# Top 10 categories for visualization
plt.figure(figsize=(14, 7))
sns.boxplot(x='category', y='stars', data=df_top10)
plt.xticks(rotation=45)
plt.title('Rating Distribution by Category (Top 10)')
plt.show()

# Highest Median Rating (No filter)
highest_rating_cat = df_clean.groupby('category')['stars'].median().idxmax()
highest_rating_val = df_clean.groupby('category')['stars'].median().max()
print(f"The category with the highest median rating is: {highest_rating_cat} ({highest_rating_val} stars)")
```

Business Observations
When writing your report, consider:

Price vs. Volume: Are the categories with the most products (Top 10) also the most expensive? (Usually, they are mid-range).

Rating Consistency: If a box plot is very "short," it means customers generally agree on the quality. If it is "tall," the quality in that category is highly inconsistent.

Part: 3

In this final part, we examine if "expensive" implies "high quality" in the eyes of Amazon UK customers. We will use the df_clean dataframe (without outliers) to ensure our correlations aren't skewed by a few luxury items.
Part 3: Investigating Price and Rating Interplay

1. Correlation AnalysisThe Pearson Correlation Coefficient (1$r$) tells us how two variables move together.

r = 1: Perfect positive relationship.
r = 0: No relationship.
r = -1: Perfect negative relationship.

```python
# Calculate correlation coefficient
correlation = df_clean['price'].corr(df_clean['stars'])

print(f"Correlation between Price and Stars: {correlation:.4f}")

if abs(correlation) < 0.1:
    print("Insight: There is a negligible correlation between price and ratings.")
elif abs(correlation) < 0.3:
    print("Insight: There is a weak correlation between price and ratings.")
else:
    print("Insight: There is a moderate to strong correlation.")
```

2. Visualizations
Scatter Plot: Price vs. Stars
This visualization helps us see if points cluster in a specific direction.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='stars', y='price', data=df_clean, alpha=0.3, color='purple')
plt.title('Scatter Plot: Product Rating vs. Price', fontsize=14)
plt.xlabel('Rating (Stars)')
plt.ylabel('Price (£)')
plt.show()
```

Correlation Heatmap: 

This checks the relationship between all numerical values (price, stars, reviews, etc.).

```python
plt.figure(figsize=(10, 8))
# Select only numerical columns for correlation
numerical_cols = df_clean.select_dtypes(include=[np.number])
sns.heatmap(numerical_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Variables', fontsize=14)
plt.show()
```



QQ Plot: Normality Test

The Quantile-Quantile (QQ) Plot is used to see if our price data follows a Normal Distribution (Bell Curve). If the blue dots fall on the red line, the data is normal.

```python
import scipy.stats as stats

plt.figure(figsize=(8, 6))
stats.probplot(df_clean['price'], dist="norm", plot=plt)
plt.title('QQ Plot: Price Distribution vs Normal Distribution', fontsize=14)
plt.show()
```

Business-Centric Findings for Part 3:
Price-Quality Gap: On Amazon, price and stars often have a very low correlation. This suggests that customers rate products based on expectations rather than absolute price. A cheap product that works well gets 5 stars, while an expensive product that is "just okay" might get 3 stars.

Price Distribution: Your QQ plot will likely show the dots curving away from the line at the ends. This indicates that even after removing outliers, Amazon prices are skewed and do not follow a perfect bell curve.

Heatmap Insights: Check if reviews correlate with stars. Often, products with more reviews have more stable (but slightly lower) average ratings.

Bonus Task: Analysis with Outliers
If you perform this same analysis without removing outliers, you will see:

Scatter Plots: All the data will be squashed at the bottom because one or two £5,000 items will scale the Y-axis so high that the £10 items become invisible.
